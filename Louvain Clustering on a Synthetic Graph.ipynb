```python
import networkx as nx
import matplotlib.pyplot as plt
from community import community_louvain

# Create a graph with planted communities
G = nx.planted_partition_graph(l=4, k=15, p_in=0.5, p_out=0.01, seed=42)

# Run Louvain community detection
partition = community_louvain.best_partition(G)

# Count communities
num_communities = len(set(partition.values()))
print(f"Detected {num_communities} communities")

pos = nx.spring_layout(G, seed=42)
colors = [partition[node] for node in G.nodes()]
nx.draw_networkx(G, pos, node_color=colors, with_labels=False, node_size=100, cmap=plt.cm.Set3)
plt.title(f'Louvain: {num_communities} communities')
plt.axis('off')
plt.show()
```
